Logestic regression (single input and output)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=pd.read_excel('random_dataset_50.xlsx')
df.drop(columns=["EstimatedSalary"],inplace=True)
df.head()
```

```python
plt.figure(figsize=(10,3))
sns.scatterplot(x="Age", y= "Purchased",data= df)
plt.show()
```

```python
x =df[["Age"]]
y = df[["Purchased"]]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y,
                                                    test_size=0.2,
                                                    random_state=20
                                                    )

print(x.shape, y.shape)
print(x_train.shape, y_train.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[25]])
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x="Age", y= "Purchased",data= df)
sns.lineplot(x="Age",y=lr.predict(x),data=df,color='red')
plt.show()
```
