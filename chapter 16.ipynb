POLYNOMIAL REGRESSION  : it is a method of regression analysis in which the relationship between the independent variable(s) and the dependent variable.

it follow random pattern and it is used to predict the value of the dependent variable based on the value of the independent

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_excel('salary_by_level_simple.xlsx')
df.head()
```

```python
df.corr()
```

```python
plt.scatter(df['Level'], df['Salary'], color='red', alpha=0.5)
plt.xlabel('Level')
plt.ylabel('Salary')

plt.title('Salary by Level')
plt.show()
```

```python
x = df['Level']
y = df['Salary']
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures

x = np.array([1, 2, 3])  # 1D array
x = x.reshape(-1, 1)     # Convert to 2D: shape (n_samples, 1)

pf = PolynomialFeatures(degree=2)
X_poly = pf.fit_transform(x)

print(X_poly)
```

```python
x = X_poly
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,
                                                    test_size=0.2,
                                                    random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
prd=model.predict(x)
```

```python
plt.scatter(df['Level'], df['Salary'], color='red', alpha=0.5)
plt.xlabel('Level')
plt.ylabel('Salary')
plt.plot(df['Level'], prd, color='blue', linewidth=2)
plt.title('Salary by Level')
plt.legend(['Actual', 'Predicted'])
plt.show()
```
