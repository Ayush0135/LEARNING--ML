Multiple linear regression :

models are used to predict the value of a continuous outcome variable based on one or more predictor variables

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("age_experience_salary_dataset.csv")
df.head()
```

```python
df.shape
df.isnull().sum()
```

```python
sns.pairplot(df, kind="reg")
```

```python
sns.heatmap(data=df.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
```

```python
x = df.iloc[:,:-1]
y = df["Salary (INR)"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

 Y = M1X1+M2X2......+C

```python
lr.score(x_test, y_test)*100
```

```python
lr.coef_
```

```python
lr.intercept_
```
